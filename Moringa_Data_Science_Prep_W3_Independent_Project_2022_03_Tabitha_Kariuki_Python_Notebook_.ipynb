## **Understanding and Cleaning Data** 

```python
# Combining the last three datasets namely; CDR 20120507 CDR 20120508 CDR 20120509
# they have similar columns
# respresented by df4 df5 df6 respectively

combined_table = pd.concat([df4, df5, df6], ignore_index=1)
 
display(combined_table)
```

```python
# Checking the new columns after combining the df4 df5 df6 to get better understanding of our new table.
# As well check for their unique values using the nunique command.

combined_table.columns

combined_table.nunique
```

```python
# Checking the main products offered by MTN Coted'Ivore

combined_table['PRODUTC'].unique()
```

```python
# Checking if there are missing values in your data

combined_table.isnull().sum()
```

```python
# Dropping all the N/A OR NULL values in our combined table.(for illustration purposes)
# But since the null values are too many I will work with the table as it is and I will repalce all NA values with 0. 

combined_table.dropna (axis = 'rows')

combined_table.dropna (axis = 'columns')
```

```python
combined_table.fillna(0)
```

```python
# Dropping rows and columns in pandas Dataframe

final_table = combined_table.drop(['CELL_ON_SITE', 'COUNTRY_A', 'COUNTRY_B'], axis=1)
```
